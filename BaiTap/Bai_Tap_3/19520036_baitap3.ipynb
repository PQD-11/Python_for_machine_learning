```python
import geopandas as gpd
import folium
import matplotlib.pyplot as plt
```

```python
!git clone https://github.com/CityScope/CSL_HCMC
```

```python
path = 'CSL_HCMC/Data/GIS/Population/population_HCMC/population_shapefile/Population_District_Level.shp'
df = gpd.read_file(path)
df.head()
```

```python
df.plot(figsize=(8, 8))
plt.show()
```

```python
df.crs
```

```python
df = df.to_crs(epsg=4326)
print(df.crs)
df.head()
```

```python
df.plot(figsize=(8, 8))
plt.show()
```

```python
m = folium.Map(location=[10.8, 106.7], zoom_start=10, tiles='CartoDB positron')
m
```

```python
for _, r in df.iterrows():
    sim_geo = gpd.GeoSeries(r['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j,
                           style_function=lambda x: {'fillColor': 'blue'})
    folium.Popup(r['Dist_Name']).add_to(geo_j)
    geo_j.add_to(m)
m
```

```python
df = df.to_crs(epsg=2263)
df['centroid'] = df.centroid
```

```python
df = df.to_crs(epsg=4326)
df['centroid'] = df['centroid'].to_crs(epsg=4326)
df.head()
```

```python
for _, r in df.iterrows():
    lat = r['centroid'].y
    lon = r['centroid'].x
    folium.Marker(location=[lat, lon],
                  popup='length: {} <br> area: {}'.format(r['Shape_Leng'], r['Shape_Area'])).add_to(m)
m
```

```python
df.plot(column='Shape_Area', cmap='Blues', edgecolor='b', figsize=(8, 8))
plt.show()
```
